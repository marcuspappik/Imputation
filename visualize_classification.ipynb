```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = pd.read_csv('data/datasets.csv')
```

```python
clf_datasets = datasets.loc[datasets['type']=='classification', 'dataset']
```

```python
for clf_dataset in clf_datasets:
    clf_results = pd.read_csv('data/'+clf_dataset+'/results_evaluation.csv')
    for p in [0.2, 0.4, 0.6, 0.8]:
        plt.figure(figsize=(16,5))
        i=1
        for mechanism in ['mcar', 'mar', 'mnar']:
            condition = (clf_results["('settings', 'mechanism')"]==mechanism) & (clf_results["('settings', 'p')"]==p)
            mice_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & condition,
                                        ["('settings', 'number')", "('f1-score', 'post')"]]
            mice_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & condition,
                                        ["('settings', 'number')", "('f1-score', 'prev')"]]
            mida_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & condition,
                                        ["('settings', 'number')", "('f1-score', 'post')"]]
            mida_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & condition,
                                        ["('settings', 'number')", "('f1-score', 'prev')"]]
            plt.subplot(1, 3, i)
            plt.title(clf_dataset+': f1-score '+mechanism+' p='+str(p))
            plt.xlabel('number of imputations')
            plt.ylabel('f1-score')
            plt.plot(mice_post["('settings', 'number')"], mice_post["('f1-score', 'post')"], label='mice post')
            plt.plot(mice_prev["('settings', 'number')"], mice_prev["('f1-score', 'prev')"],  label='mice prev')
            plt.plot(mida_post["('settings', 'number')"], mida_post["('f1-score', 'post')"],  label='mida post')
            plt.plot(mida_prev["('settings', 'number')"], mida_prev["('f1-score', 'prev')"],  label='mida prev')
            i += 1
        plt.legend()
        plt.show()
```

```python
for clf_dataset in clf_datasets:
    clf_results = pd.read_csv('data/'+clf_dataset+'/results_evaluation.csv')

    for mechanism in ['mcar', 'mar', 'mnar']:
        knn = clf_results.loc[(clf_results["('settings', 'imputation')"]=='KNN') &
                              (clf_results["('settings', 'mechanism')"]==mechanism), 
                              ["('settings', 'p')", "('f1-score', 'prev')"]] 
        soft = clf_results.loc[(clf_results["('settings', 'imputation')"]=='SoftImpute') &
                               (clf_results["('settings', 'mechanism')"]==mechanism), 
                               ["('settings', 'p')", "('f1-score', 'prev')"]] 

        mice_1_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & 
                                      (clf_results["('settings', 'number')"]==1) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'prev')"]]  
        mice_1_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & 
                                      (clf_results["('settings', 'number')"]==1) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'post')"]] 


        mice_20_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & 
                                      (clf_results["('settings', 'number')"]==20) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'prev')"]] 
        mice_20_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & 
                                      (clf_results["('settings', 'number')"]==20) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'post')"]] 


        mice_40_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & 
                                       (clf_results["('settings', 'number')"]==40) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                       ["('settings', 'p')", "('f1-score', 'prev')"]] 
        mice_40_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & 
                                       (clf_results["('settings', 'number')"]==40) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                       ["('settings', 'p')", "('f1-score', 'post')"]] 

        mida_1_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & 
                                      (clf_results["('settings', 'number')"]==1) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'prev')"]]  
        mida_1_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & 
                                      (clf_results["('settings', 'number')"]==1) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'post')"]] 


        mida_20_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & 
                                      (clf_results["('settings', 'number')"]==20) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'prev')"]] 
        mida_20_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & 
                                      (clf_results["('settings', 'number')"]==20) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                      ["('settings', 'p')", "('f1-score', 'post')"]] 


        mida_40_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & 
                                       (clf_results["('settings', 'number')"]==40) &
                                       (clf_results["('settings', 'mechanism')"]==mechanism), 
                                       ["('settings', 'p')", "('f1-score', 'prev')"]] 
        mida_40_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & 
                                       (clf_results["('settings', 'number')"]==40) &
                                      (clf_results["('settings', 'mechanism')"]==mechanism), 
                                       ["('settings', 'p')", "('f1-score', 'post')"]] 
        
        
        plt.figure(figsize=(16,5))
        plt.subplot(1, 2, 1)
        plt.title('f1-score for differet p with prev')
        plt.xlabel('p')
        plt.ylabel('f1-score')
        plt.plot(knn["('settings', 'p')"], knn["('f1-score', 'prev')"], label='knn')
        plt.plot(soft["('settings', 'p')"], soft["('f1-score', 'prev')"], label='softimpute')
        # plt.plot(mice_1_prev["('settings', 'p')"], mice_1_prev["('f1-score', 'prev')"], label='mice 1')
        plt.plot(mice_20_prev["('settings', 'p')"], mice_20_prev["('f1-score', 'prev')"], label='mice 20 prev')
        plt.plot(mice_40_prev["('settings', 'p')"], mice_40_prev["('f1-score', 'prev')"], label='mice 40 prev')
        # plt.plot(mida_1_prev["('settings', 'p')"], mida_1_prev["('f1-score', 'prev')"], label='mida 1')
        plt.plot(mida_20_prev["('settings', 'p')"], mida_20_prev["('f1-score', 'prev')"], label='mida 20 prev')
        plt.plot(mida_40_prev["('settings', 'p')"], mida_40_prev["('f1-score', 'prev')"], label='mida 40 prev')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.title('f1-score for differet p with post')
        plt.xlabel('p')
        plt.ylabel('f1-score')
        plt.plot(knn["('settings', 'p')"], knn["('f1-score', 'prev')"], label='knn')
        plt.plot(soft["('settings', 'p')"], soft["('f1-score', 'prev')"], label='softimpute')
        # plt.plot(mice_1_post["('settings', 'p')"], mice_1_post["('f1-score', 'post')"], label='mice 1')
        plt.plot(mice_20_post["('settings', 'p')"], mice_20_post["('f1-score', 'post')"], label='mice 20 post')
        plt.plot(mice_40_post["('settings', 'p')"], mice_40_post["('f1-score', 'post')"], label='mice 40 post')
        # plt.plot(mida_1_post["('settings', 'p')"], mida_1_post["('f1-score', 'post')"], label='mida 1')
        plt.plot(mida_20_post["('settings', 'p')"], mida_20_post["('f1-score', 'post')"], label='mida 20 post')
        plt.plot(mida_40_post["('settings', 'p')"], mida_40_post["('f1-score', 'post')"], label='mida 40 post')

        plt.legend()
        plt.show()
```
