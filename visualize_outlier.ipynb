```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
classifier = 'IsolationForest'
clf_dataset='wine'
```

```python
datasets = pd.read_csv('data/datasets.csv')
```

```python
clf_datasets = datasets.loc[datasets['type']=='outlier', 'dataset']
```

```python
clf_results = pd.read_csv('data/'+clf_dataset+'/results_'+classifier+'.csv')
for p in [0.2, 0.4, 0.6, 0.8]:
    plt.figure(figsize=(16,5))
    i=1
    for mechanism in ['mcar', 'mar', 'mnar']:
        condition = (clf_results["('settings', 'mechanism')"]==mechanism) & (clf_results["('settings', 'p')"]==p)
        mice_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & condition,
                                    ["('settings', 'number')", "('auc', 'post')"]]
        mice_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mice') & condition,
                                    ["('settings', 'number')", "('auc', 'prev')"]]
        mida_post = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & condition,
                                    ["('settings', 'number')", "('auc', 'post')"]]
        mida_prev = clf_results.loc[(clf_results["('settings', 'imputation')"]=='Mida') & condition,
                                    ["('settings', 'number')", "('auc', 'prev')"]]
        plt.subplot(1, 3, i)
        plt.title(clf_dataset+': auc '+mechanism+' p='+str(p))
        plt.xlabel('number of imputations')
        plt.ylabel('auc')
        plt.plot(mice_post["('settings', 'number')"], mice_post["('auc', 'post')"], label='mice post')
        plt.plot(mice_prev["('settings', 'number')"], mice_prev["('auc', 'prev')"],  label='mice prev')
        plt.plot(mida_post["('settings', 'number')"], mida_post["('auc', 'post')"],  label='mida post')
        plt.plot(mida_prev["('settings', 'number')"], mida_prev["('auc', 'prev')"],  label='mida prev')
        i += 1
    plt.legend()
    plt.show()
```
